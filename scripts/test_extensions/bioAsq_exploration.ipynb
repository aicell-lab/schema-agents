The snippets always come from the documents listed in the same question entry

```python
def snippets_from_documents(entry):
    document_urls = set(entry['documents'])
    for snippet in entry['snippets']:
        if snippet['document'] not in document_urls:
            return False
    return True
[x for x in [snippets_from_documents(x) for x in taskb_data['questions']] if not x]
```
