```python
image_location = '/home/alalulu/workspace/flask-chat-app-article/bioimage.png'
from PIL import Image

def load_image(image_location: str) -> Image:
    try:
        image = Image.open(image_location)
        return image
    except Exception as e:
        print(f"Error loading image: {e}")
        return None
    
def test_load_image(image_location):
    image_location = image_location
    image = load_image(image_location)
    assert image is not None, "Image loading failed"
    assert isinstance(image, Image.Image), "Invalid image object"
    print("load_image function passed all tests")

test_load_image(image_location)
load_image(image_location)
```

```python
from PIL import Image

def preprocess_image(image: Image, cell_color: str, background_color: str) -> Image:
    try:
        # Convert the image to RGB mode if it's not already
        if image.mode != "RGB":
            image = image.convert("RGB")
        
        # Create a new image with the same size and mode as the original image
        preprocessed_image = Image.new(image.mode, image.size)
        
        # Iterate over each pixel in the image
        for x in range(image.width):
            for y in range(image.height):
                # Get the RGB values of the current pixel
                r, g, b = image.getpixel((x, y))
                
                # Check if the pixel color matches the cell color
                if (r, g, b) == cell_color:
                    # Set the pixel color to the background color
                    preprocessed_image.putpixel((x, y), background_color)
        
        return preprocessed_image
    except Exception as e:
        print(f"Error preprocessing image: {e}")
        return None
    
def test_preprocess_image():
    image_location = "path/to/image.jpg"
    image = load_image(image_location)
    assert image is not None, "Image loading failed"
    
    cell_color = (0, 255, 0)  # Green color
    background_color = (255, 255, 255)  # White color
    
    preprocessed_image = preprocess_image(image, cell_color, background_color)
    assert preprocessed_image is not None, "Image preprocessing failed"
    assert isinstance(preprocessed_image, Image.Image), "Invalid preprocessed image object"
    
    print("preprocess_image function passed all tests")

test_preprocess_image()
```

```python
from PIL import Image

def count_cells(image: Image) -> int:
    try:
        # Convert the image to RGB mode if it's not already
        if image.mode != "RGB":
            image = image.convert("RGB")
        
        # Initialize a counter for the cells
        cell_count = 0
        
        # Iterate over each pixel in the image
        for x in range(image.width):
            for y in range(image.height):
                # Get the RGB values of the current pixel
                r, g, b = image.getpixel((x, y))
                
                # Check if the pixel color is not the background color (white)
                if (r, g, b) != (255, 255, 255):
                    # Increment the cell count
                    cell_count += 1
        
        return cell_count
    except Exception as e:
        print(f"Error counting cells: {e}")
        return 0
    
def test_count_cells():
    image_location = "path/to/image.jpg"
    image = load_image(image_location)
    assert image is not None, "Image loading failed"
    
    preprocessed_image = preprocess_image(image, (0, 255, 0), (255, 255, 255))
    assert preprocessed_image is not None, "Image preprocessing failed"
    
    cell_count = count_cells(preprocessed_image)
    assert cell_count == 10, "Incorrect cell count"
    
    print("count_cells function passed all tests")

test_count_cells()
```
